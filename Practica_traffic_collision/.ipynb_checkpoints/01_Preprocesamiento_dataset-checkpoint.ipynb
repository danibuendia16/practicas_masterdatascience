# PREPROCESAMIENTO DEL DATASET

```python
# carga de las librerías necesarias
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.preprocessing import OneHotEncoder
import plotly.express as px
```

```python
# Declaración de las funciones utilizadas
def transform_hour(value):
    '''
    La siguiente función transforma en franjas horarias, las distintas horas de un día. Define los siguientes grupos:
    {0: madrugada (de 00:00-6:00), 1: mañana (6:00-12:00), 2: tarde (12:00-18:00), 3: tarde-noche (18:00-24:00)}
    '''
    value = pd.to_numeric(value,errors='coerce')
    if value >= 0 and value<6:
        return '0'
    if value >= 6 and value<12:
        return '1'
    if value >= 12 and value<18:
        return '2'
    if value >= 18 and value<24:
        return '3'
    return pd.NA
```

```python
# Lectura de los datos
data = pd.read_csv('data/pd_data_EDA_missing_outlier.csv', low_memory=False) # dtype=dtypes, 
data
```

## 00. CREACIÓN DE VARIABLES AUXILIARES

El dataset está diseñado inicialmente de tal forma que cada registro hace referencia a cada persona involucrada en los distintos accidentes. Es decir, el dataset está "agrupado" por personas. Sin embargo, considerando el propóstio de este trabajo detallado anteriormente, esto no tiene mucho sentido, ya que nos interesa conocer los datos agregados por accidente de cara a conocer la posible mortalidad en los mismos. Con este propósito se va a transformar el dataset, modificando y agrupando las variables pertenecientes a niveles inferiores al de accidentes, de tal forma que el dataset sobre el que se va a modelizar este "agrupado" por accidentes.

### Variable P_SAFE

A la hora de tratar la variable P_SAFE se ha determinado que, para simplificar la información que aporta al modelo, se va a diferenciar únicamente en si se había utilizado algún tipo de sistema de seguridad en el accidente o no. Se considera que en aquellos casos en los que el elemento no es aplicable o que el valor es distinto al del resto de grupos, la persona sí disponía de un elemento de seguridad.
Por último, para que la variable tenga un mayor sentido a nivel agregado, se estimará una vez se agrupe el dataset por accidente como el porcentaje de individuos involucrados que disponían de un dispositivo de seguridad respecto el total de individuos implicados en el accidente. Para ello, se comparará con el total de individuos involucrados en cada accidente.

```python
# Transformamos las distintas categorías de la variable original, simplificándola
dict_psafe = {
        1: 0,
        2: 1,
        9: 1,
        10: 1,
        11: 1,
        12: 1,
        13: 0,
        14: 1,
}

data = data.replace({'P_SAFE': dict_psafe})
data
```

Cuando las variables se agrupen por accidentes en el dataframe definitivo para la modelización, esta variable se transformará como porcentaje sobre el total de individuos implicados en cada accidente.

```python
# definición de la tipología de las variables tras la carga de datos
target = ['C_SEV']
numeric_features = ['C_YEAR', 'C_VEHS', 'P_AGE', 'P_SAFE', 'V_YEAR']
categories = [col for col in data.columns if col not in target+numeric_features]
```

```python
dtypes = {}
for feature in numeric_features:
    dtypes[feature] = 'float' 
for feature in categories+target:
    dtypes[feature] = 'str' 
```

```python
data = data.astype(dtypes)
```

```python
data.dtypes.to_dict()
```

### Variables C_YEAR y V_YEAR

La siguiente transformación a realizar será determinar la antigüedad del vehículo implicado en la colisión mediante la fecha de colisión y la fecha de antigüedad del vehículo. Poseriormente dichas columnas serán eliminadas, ya que tanto el año del accidente como el año de fabricación del vehículo realmente no son un factor determinante en una colisión, mientras que la nueva variable estimada sí puede aportar información adicional sobre el accidente.

```python
# creamos una columna que nos informe de la antigüedad del vehículo 
data['A_VAGE'] = data['C_YEAR'] - pd.to_numeric(data['V_YEAR'], errors='coerce')
data
```

Como se acaba de comentar, C_YEAR y V_YEAR serán eliminadas del dataset. C_YEAR, sin embargo, se eliminará tras hacer la agregación ya que es necesaria para diferenciar los distintos accidentes que se han producido durante el periodo de estudio.

```python
data = data.drop('V_YEAR', axis=1)
data
```

### Variable P_PSN

Se considera que esta variable recoge la misma información ya contenida en la variable P_USER a un nivel de detalle mayor, que no aporta una utilidad significativa al modelo, por lo que será eliminada.

```python
data = data.drop('P_PSN', axis=1)
data
```

## 01. ENCODING DE LAS VARIABLES NO ACCIDENTES

Con el objetivo de poder agrupar el dataset por accidente, se deberá realizar de manera independiente una codificación de aquellas variables categóricas que incluyen datos desagregados por vehículo o persona (es decir distintas a 'C_XXXX').
Para poder agrupar posteriormente los datos con sentido, esta codificación se realizará utilizando OneHot encoding.

```python
target = ['C_SEV']
numeric_features = ['C_YEAR', 'C_VEHS', 'A_VAGE','P_AGE', 'P_SAFE']
categories = [col for col in data.columns if col not in target+numeric_features]
```

```python
c_cols = [col for col in data.columns if col[0]=="C"]
```

```python
categories = [col for col in data.columns if col not in target+numeric_features]
categories_onehot_encoding = []
for i in categories:
    if i not in c_cols:
        categories_onehot_encoding.append(i)

categories_onehot_encoding
```

```python
data = pd.get_dummies(data, columns=categories_onehot_encoding,
                      prefix=categories_onehot_encoding) #, dtype=int
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
change_type = [col for col in data.columns if col not in c_cols and data[col].dtype == np.uint8]
change_type.append('P_SAFE')
```

## 02. AGREGACIÓN DEL DATASET POR ACCIDENTE

```python
# definición de las funciones de agregación que se utilizarán
agg_dict = {}
for feature in change_type:
    agg_dict[feature] = 'sum' 
agg_dict['P_AGE'] = np.median
agg_dict['A_VAGE'] = np.mean
agg_dict
```

```python
# Agregamos los datos en función de estas columnas, agrupando por accidentes
data = data.groupby(c_cols).agg(agg_dict).reset_index()
data
```

## 03. Creación de variables auxiliares tras la agrupación

### Variable P_SAFE

Como se ha descrito anteriormente, con el objetivo de incrementar el valor de la variable, se va a estimar como porcentaje sobre el total de individuos involucrados en el accidente creándose una nueva variable A_PSAF que recoja esta información y eliminándose la anterior P_SAFE. Asimismo, se eliminará la variable C_YEAR ya que se considera que no aporta información relevante al modelo.

```python
data['P_SAFE'] = data['P_SAFE']/(data['P_SEX_F'] + data['P_SEX_M'])
data = data.drop('C_YEAR', axis=1)
data
```

### Variable C_HOUR

A continuación, vamos a transformar la variable que recoge información sobre la hora en la que se produjeron las distintas colisiones. Originalmente, esta variable recoge las horas en las que se produjeron los accidentes, sin embargo, se considera que agregando la información en tramos horarios, puede aportar un mayor valor al modelo que considerando de manera independiente cada hora. Las categorías generadas son las siguientes:
* Madrugada (de 00:00-6:00)
* Mañana (6:00-12:00)
* Tarde (12:00-18:00)
* Tarde-noche (18:00-24:00)

Para ello se ha utilizado la función definida con este propósito, en el apartado correspondiente

```python
# Aplicamos la función transdorm_hour a la columna de la hora de colisión
data['C_HOUR'] = data['C_HOUR'].apply(transform_hour) 
data
```

```python
data.dtypes.to_dict()
```

```python
change_type = [col for col in data.columns if data[col].dtype == float]
change_type.remove('P_SAFE')
data[change_type] = data[change_type].astype(np.uint8)
```

```python
data
```

## 04. Correlación variables

Antes de comenzar el modelado, como último paso previo, se va a representar mediante una matriz de correlaciones, entre las variables numéricas, ya fueran tanto numéricas en origen como las categóricas que se han codificado para poder agrupar el dataset por accidente.
Debido a que existen variables de naturaleza tanto numérica como categórica, el método utilizado será la correlación de Spearman.

```python
list_numeric_var = list(data.select_dtypes(['float', 'uint8']).columns)
sns.set(style="white")
# Calcular la matriz de correlaciones
corr = data[list_numeric_var].corr(method='spearman') 
# Representación gráfica
f, ax = plt.subplots(figsize=[10,8])
# Draw the heatmap with the mask and correct aspect ratio
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
    
ax = sns.heatmap(corr, mask=mask, center=0,
                 square=True, linewidths=.5,  cmap ='plasma' ) #cbar_kws={"shrink": .5}
plt.show()
```

Como se puede observar en el gráfico, al agrupar por accidente, las variables P_USER y V_TYPE, tienden a solaparse en cuanto a la información que contienen, provocando en aquellos valores que comparten que se encuentren dupicados como en el caso de motoristas y ciclistas.
Por este motivo, se ha decidido eliminar la variable P_USER ya que contiene menos información que puede resultar ser relevante para el modelo.

```python
p_user_col = [col for col in data.columns if col[0:6]=="P_USER"]
data = data.drop(p_user_col, axis=1)
data
```
